```python
root_path = '/home2/dataset/Empathetic_Dataset/MSC/'
```

## Add video feature path

```python
video_path_list = []

for index, row in csv.iterrows():
    d = row['Dialogue_ID']
    u = row['Utterance_ID']
    file_path = root_path + f'{mode}/video_feature/dia{d}_utt{u}.pt'

    if os.path.exists(file_path):
        video_path_list.append(file_path)
    else:
        print(f"The file {file_path} doesn't exists. (idx={index})")
        video_path_list.append(None)

csv['video_feature_path'] = video_path_list

csv
```
